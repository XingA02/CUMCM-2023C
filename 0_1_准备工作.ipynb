```python
df3 = pd.read_csv("中间数据集\\附件3的csv形式.csv")
df3['日期'] = pd.to_datetime(df3['日期'])
df3.head()
```

```python
df4 = pd.read_excel("0_赛题\\附件4.xlsx", sheet_name="Sheet1").drop("单品编码", axis=1)
df4.columns = ['单品名称', '损耗率_单品']
df4.head()
```

```python
df4__ = pd.read_excel("0_赛题\\附件4.xlsx").drop("小分类编码", axis=1)
df4__.columns = ['分类名称', '损耗率_品类']
df4__.head()
```

## ***数据集成***

```python
df_a = pd.merge(df1, df2, on="单品编码", how="right").drop("分类编码", axis=1)
df_a
```

```python
df_b = pd.merge(df_a, df3, on=['日期', '单品编码'], how="left")
df_b
```

```python
df_c = pd.merge(df_b, df4, on=['单品名称'], how="left")
df_c.head()
```

```python
df_combined = pd.merge(df_c, df4__, on=['分类名称'], how="left")
df_combined.head()
```

```python
df_combined.columns
```

```python
df_combined_final = df_combined[['日期', '单品编码', '单品名称', '分类名称', '销量(千克)', 
                        '销售单价(元/千克)', '销售类型', '是否打折销售',  
                        '批发价格(元/千克)', '损耗率_单品', '损耗率_品类']]
df_combined_final.to_csv("中间数据集\\超级大表.csv", index=None)
```

## ***数据划分***

```python
df_combined_final.head()
```

```python
# 假设df是你的数据集DataFrame，category_column是分类列的名称
# 首先，找出所有不同的类别
categories = df_combined_final["分类名称"].unique()

# 然后，为每个类别创建一个空的DataFrame
category_datasets = {category: pd.DataFrame() for category in categories}

# 遍历原始数据集，并将每个记录分配到相应的类别DataFrame中
for category in categories:
    splite = df_combined_final[df_combined_final["分类名称"] == category]
    splite = splite.drop('损耗率_单品', axis=1)
    splite.to_csv(f"中间数据集\\多表_按品类\\{category}.csv", index=None)
```
